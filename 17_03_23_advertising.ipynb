```python
import pandas as pd
url = 'Advertising.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['TV']),y=data['sales'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri coeffient la :',model.coef_)
```

```python
X=pd.DataFrame(data['TV'])
y=data['sales']
model.score(X,y)
```

```python
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black')
y_pred=model.predict(X)
plt.plot(data['TV'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['TV'],data['TV']**2]).T
model.fit(X,y=data['sales'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la :',model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X,y))
```

```python
import numpy as np
range=np.arange(100,200,1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X_range)
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black',xlim=(0.7))
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3,data['TV']**4,data['TV']**5,data['TV']**6,data['TV']**7,data['TV']**8,data['TV']**9,data['TV']**10]).T
model.fit(X,y=data['sales'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la :',model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X,y))
```

```python
import numpy as np
range=np.arange(100,200,1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred=model.predict(X_range)
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black',xlim=(0.7))
plt.plot(range,y_pred,color='blue')
```
